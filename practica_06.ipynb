## Ejercicio 2 — Calculadora básica

```scala
val a: Double = 15.0
val b: Double = 4.0

val suma     = a + b
val resta    = a - b
val producto = a * b
val division = a / b
val modulo   = a % b

println("=== Calculadora Scala ===")
println(f"  $a%.1f + $b%.1f = $suma%.1f")
println(f"  $a%.1f - $b%.1f = $resta%.1f")
println(f"  $a%.1f × $b%.1f = $producto%.1f")
println(f"  $a%.1f ÷ $b%.1f = $division%.4f")
println(f"  $a%.1f %% $b%.1f = $modulo%.1f")
```

```scala
// División entera con Int
val x: Int = 15
val y: Int = 4

println(s"División entera: $x / $y = ${x / y}")
println(s"Resto:           $x % $y = ${x % y}")

// Comparaciones
println(s"\n¿$a > $b?   ${a > b}")
println(s"¿$a == $b?  ${a == b}")
println(s"¿$a != $b?  ${a != b}")
```

```scala
val pesoKg    = 87.0
val alturaMts = 1.74

val imc = pesoKg / (alturaMts * alturaMts)

val categoria =
  if      (imc < 18.5) "Bajo peso"
  else if (imc < 25.0) "Normal"
  else if (imc < 30.0) "Sobrepeso"
  else                 "Obesidad"

println(f"IMC: $imc%.2f → $categoria")
```

Ejercicio P10 — Conversor de unidades múltiple
Declara val distanciaKm: Double = 142.5. A partir de solo esa variable, calcula la equivalencia en las cuatro unidades siguientes y muestra la tabla con el interpolador f:

```scala
val distanciaKm: Double = 142.5

val metrosPorKm = 1000.0
val millasPorKm = 0.621371
val piesPorKm = 3280.84
val pulgadasPorKm = 39370.1

val enMetros = distanciaKm * metrosPorKm
val enMillas = distanciaKm * millasPorKm
val enPies = distanciaKm * piesPorKm
val enPulgadas = distanciaKm * pulgadasPorKm

println(f"=== Conversor de distancia ===")
println(f"Distancia original: ${distanciaKm}%10.2f km")
println()

println(f"En metros:   ${enMetros}%12.2f m")
println(f"En millas:   ${enMillas}%12.2f millas")
println(f"En pies:     ${enPies}%12.2f pies")
println(f"En pulgadas: ${enPulgadas}%12.2f pulgadas")
```
